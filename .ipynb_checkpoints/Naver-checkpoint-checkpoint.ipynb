```python
id_list=[]
place_list=[]
tel_list=[]
address_list=[]
```

```python
searchtext = '권선구인테리어 디자인'
nmap_url = 'https://m.map.naver.com/search2/search.naver?query='+searchtext
driver.get(nmap_url)
time.sleep(0.5)  # allow the webpage to load fully before further interaction.
```

```python
for i in range(1,76):#關鍵字顯示比數
    #HTML 自訂義屬性
    #//*[@id="ct"]/div[2]/ul/li[1]
    xpath_id = driver.find_element("xpath",'/html/body/div[4]/div[2]/ul/li['+str(i)+']').get_attribute("data-id")
    id_list.append(xpath_id)
    
    xpath_title = driver.find_element("xpath",'/html/body/div[4]/div[2]/ul/li['+str(i)+']').get_attribute("data-title")
    place_list.append(xpath_title)
    
    xpath_tel = driver.find_element("xpath",'/html/body/div[4]/div[2]/ul/li['+str(i)+']').get_attribute("data-tel")
    tel_list.append(xpath_tel)
    
    xpath_address = driver.find_element("xpath",'/html/body/div[4]/div[2]/ul/li['+str(i)+']/div[1]/div[1]/div/a').text
    address_list.append(xpath_address[5:])
```

```python
df = pd.DataFrame({'id':id_list,'name':place_list,'tel':tel_list,'address':address_list})
df.loc[(df['tel'].notnull())& (df['tel']!='')]
```

```python
city_district = pd.read_excel('京畿道行政區列表1006.xlsx', sheet_name = '京畿道行政區列表', usecols = ['Unnamed: 1', 'Unnamed: 2'], dtype='object')
```

```python
key_word = pd.read_excel('京畿道行政區列表1006.xlsx', sheet_name = '京畿道行政區列表', usecols = ['關鍵字', '不要撈的關鍵字'], dtype='object')
```

```python
print(city_district.head())
```

```python
city_district.rename(columns={'Unnamed: 1': "City",
                            'Unnamed: 2': "District"}, inplace=True)
```

```python
key_word.rename(columns={'關鍵字': "key_word",
                        '不要撈的關鍵字': "none_key"}, inplace=True)
```

```python
print(city_district.head())
```

```python
# Define a function to remove non-desired characters
def remove_non_desired_characters(text):
    if text is not None:
        # Define a regular expression pattern to match any characters that are not in the desired set
        pattern = re.compile(f'[^가-힣]')  # This pattern matches any characters that are not Korean

        # Use sub to replace matched characters with an empty string
        return pattern.sub('', text)
    return text

city_district['City'] = city_district['City'].fillna('')
city_district['District'] = city_district['District'].fillna('')
# Apply the custom function to 'City' and 'District' columns
city_district['City'] = city_district['City'].apply(remove_non_desired_characters)
city_district['District'] = city_district['District'].apply(remove_non_desired_characters)

# Drop rows with empty 'City' or 'District'
city_district = city_district.dropna(subset=['City', 'District'])
```

```python
#print(city_district.head(10))
```

```python
city_district['City'] = city_district['City'].replace('', pd.NA).fillna(method='ffill')
```

```python
city_district= city_district[city_district['District'] != ' ']
city_district.reset_index(drop=True, inplace=True)
```

```python
city_district['combined'] = city_district['City']+ city_district['District']
```

```python
combined_results = [str(area) + str(keyword) for area in city_district['combined'] for keyword in key_word['key_word'] if not pd.isna(keyword)]
```

```python
print(combined_results)
```

```python
index = combined_results.index('포천시포천동공간 인테리어 디자인')
print(index)
# Create a new list starting from the target word
combined_results = combined_results[index:]
```

```python
print(combined_results)
```

```python
id_list=[]
place_list=[]
tel_list=[]
address_list=[]
```

```python
for searchtext in combined_results:
    nmap_url = 'https://m.map.naver.com/search2/search.naver?query=' + searchtext
    driver.get(nmap_url)
    time.sleep(2.5)

    # Set implicit wait
    driver.implicitly_wait(2.5)  # Adjust the wait time as needed

    try:
        elements = driver.find_elements(By.XPATH, '/html/body/div[4]/div[2]/ul/li')
        max_elements = min(len(elements), 75)

        for j in range(1, max_elements + 1):
            try:
                xpath_id = driver.find_element(By.XPATH, f'/html/body/div[4]/div[2]/ul/li[{j}]').get_attribute("data-id")
                id_list.append(xpath_id)
                xpath_title = driver.find_element(By.XPATH, f'/html/body/div[4]/div[2]/ul/li[{j}]').get_attribute("data-title")
                place_list.append(xpath_title)
                xpath_tel = driver.find_element(By.XPATH, f'/html/body/div[4]/div[2]/ul/li[{j}]').get_attribute("data-tel")
                tel_list.append(xpath_tel)
                xpath_address = driver.find_element(By.XPATH, f'/html/body/div[4]/div[2]/ul/li[{j}]/div[1]/div[1]/div/a').text
                address_list.append(xpath_address[5:])
            except NoSuchElementException:
                # Handle the exception here (e.g., print an error message)
                print(f"Element not found for index {j}")

    except NoSuchElementException:
        # Handle the exception here (e.g., print an error message)
        print("Elements not found on the page")
```
